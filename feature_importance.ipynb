```python
import matplotlib.pyplot as plt
import pandas as pd

importance = model.get_feature_importances()

# Plot each variable group
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, (group_name, scores) in zip(axes, importance.items()):
    # Convert to DataFrame
    df = pd.Series(scores).sort_values(ascending=True)
    
    df.plot(kind='barh', ax=ax, color='steelblue')
    ax.set_title(f'{group_name}', fontsize=13)
    ax.set_xlabel('Importance Score')
    ax.axvline(x=1/len(df), color='red', linestyle='--', label='Uniform baseline')
    ax.legend()

plt.suptitle('TFT Feature Importance (Variable Selection Networks)', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Get attention weights — shows which past timesteps the model focuses on
attention = model.get_attention()

import matplotlib.pyplot as plt
import numpy as np

# Average attention across all heads and series
avg_attention = np.mean(attention, axis=0)

plt.figure(figsize=(10, 5))
plt.plot(avg_attention)
plt.xlabel("Past Timesteps (older → recent)")
plt.ylabel("Attention Weight")
plt.title("TFT Temporal Attention — Which timesteps matter most?")
plt.grid(True)
plt.show()
```
